```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import pickle
from data_helper import pde_exact

# Add source and data file paths
import os
path = os.path.abspath(os.getcwd())
```

$u_t = Du_{xx} + ku(1-u), t \times x \in [0,1] \times [-1,1]$ <br>
Discretized equation with implicit Backward-time Center-space (BTCS) scheme: $\frac{u^{n+1}_{i}-u^{n}_{i}}{\Delta t} - ku^{n+1}_{i}(1-u^{n+1}_{i}) - \frac{D(u^{n+1}_{i-1}-2u^{n+1}_{i}+u^{n+1}_{i+1})}{\Delta x^2} = 0 $ <br>

Solve the equation: $a_i u^{n+1}_{i-1} + b_i u^{n+1}_i + c_i u^{n+1}_{i+1} = d_i$ <br>
where <br>
$ a_i = -D\Delta t$ <br>
$ b_i = \Delta x^2 -\Delta x^2 \Delta t k (1-(2u^n_i-u^{n-1}_i)) + 2 D \Delta t $ <br>
$ c_i = -D\Delta t $ <br>
$ d_i = \Delta x^2 u^n_i $ <br>

```python
def rand_coeff():
    n = 5
    c_i = np.random.rand(n)
    c1 = c_i[0]; c2 = c_i[1]; c3 = c_i[2]; c4 = c_i[3]; c5 = c_i[4];
    return c1, c2, c3, c4, c5

exact = pde_exact()
```

```python
N = 512; T = 256; num = 20

# Initial condtions
u0_ls = [];
for i in range(num):
    c1, c2, c3, c4, c5 = rand_coeff()
    u0_vals = exact.diffreac_1d_IC(c1, c2, c3, c4, c5, N)
    u0_ls.append(u0_vals);
```

```python
def fisher_solver(u0, fidelity_x,fidelity_t):
    D = 0.1; k = 0.1;
    x = np.linspace(-1,1,fidelity_x)
    t = np.linspace(0,1,fidelity_t)
    u = np.zeros((fidelity_t+1,fidelity_x))
    u[0,:] = u0
    u[1,:] = u0
    dx = x[1]-x[0]
    dt = t[1]-t[0] 
    for n in range(1,fidelity_t): # temporal loop
        a = np.zeros(fidelity_x); b = np.zeros(fidelity_x); c = np.zeros(fidelity_x); d = np.zeros(fidelity_x)
        for i in range(0,fidelity_x): # spatial loop
            # Create vectors for a, b, c, d
            a[i] = -D*dt
            b[i] = dx**2 - dx**2 * dt * k * (1-(2*u[n,i]-u[n-1,i])) + 2 * dt * D
            c[i] = -D*dt
            d[i] = (dx**2)*u[n,i]
        # BC's
        a[0] = 0; b[0] = 1; c[0] = 0; d[0] = 0 # Dirichlet Condition
        a[-1] = 0; b[-1] = 1; c[-1] = 0; d[-1] = 0
        # Solve
        u[n+1,:] = thomas_alg(a,b,c,d)
    v = u[1:,:]
    return v, x, t

def thomas_alg(a, b, c, d):
    n = len(b)
    x = np.zeros(n)
    for k in range(1,n):
        q = a[k]/b[k-1]
        b[k] = b[k] - c[k-1]*q
        d[k] = d[k] - d[k-1]*q
    q = d[n-1]/b[n-1]
    x[n-1] = q
    for k in range(n-2,-1,-1):
        q = (d[k]-c[k]*q)/b[k]
        x[k] = q
    return x
```

```python
u_ls = np.zeros((num, T, N))

for i in range(num):
    print(i)
    u0 = u0_ls[i]
    u, x, t = fisher_solver(u0, N, T)
    u_ls[i,:,:] = u

xf, tf = np.meshgrid(x, t)
```

```python
# Plot
plt.figure(figsize = (20,5))
plt.rcParams['text.usetex'] = True;

for i in range(num):
    plt.subplot(1,3,1);
    plt.plot(x, u_ls[i,0,:]);
    plt.subplot(1,3,2);
    plt.plot(x, u_ls[i,int(T/2),:]);
    plt.subplot(1,3,3);
    plt.plot(x, u_ls[i,-1,:]);
    
plt.subplot(1,3,1);
plt.ylabel(r'$u$', fontsize=20); plt.xlabel(r'$x$', fontsize=20); plt.title('Initial Condition', fontsize=20);
plt.xticks(fontsize=15); plt.yticks(fontsize=15); 
plt.subplot(1,3,2);
plt.ylabel(r'$u$', fontsize=20); plt.xlabel(r'$x$', fontsize=20); plt.title(r'$t=0.5$ snapshot', fontsize=20);
plt.xticks(fontsize=15); plt.yticks(fontsize=15);
plt.subplot(1,3,3);
plt.ylabel(r'$u$', fontsize=20); plt.xlabel(r'$x$', fontsize=20); plt.title(r'$t=1$ snapshot', fontsize=20);
plt.xticks(fontsize=15); plt.yticks(fontsize=15);

plt.savefig(path + '\\diffreac_data', dpi = 300, bbox_inches='tight')
```
