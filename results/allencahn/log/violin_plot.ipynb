```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

import os
path = os.path.abspath(os.getcwd())
os.chdir("../..")
parent_path = os.path.abspath(os.getcwd())

plt.rcParams['text.usetex'] = True
```

```python
# MH-PINN

with open(path + '\\hydra_sine_log_dict.pickle', 'rb') as handle:
    hydra_sine = pickle.load(handle)
    
with open(path + '\\hydra_sine_reg_log_dict.pickle', 'rb') as handle:
    hydra_sine_reg = pickle.load(handle)
    
with open(path + '\\hydra_tanh_log_dict.pickle', 'rb') as handle:
    hydra_tanh = pickle.load(handle)
    
with open(path + '\\hydra_tanh_reg_log_dict.pickle', 'rb') as handle:
    hydra_tanh_reg = pickle.load(handle)
    
# PI-DON

with open(path + '\\don_sine_log_dict.pickle', 'rb') as handle:
    don_sine = pickle.load(handle)
    
with open(path + '\\don_sine_reg_log_dict.pickle', 'rb') as handle:
    don_sine_reg = pickle.load(handle)
    
with open(path + '\\don_tanh_log_dict.pickle', 'rb') as handle:
    don_tanh = pickle.load(handle)
    
with open(path + '\\don_tanh_reg_log_dict.pickle', 'rb') as handle:
    don_tanh_reg = pickle.load(handle)
```

```python
fig, ax = plt.subplots(figsize=(18, 7))

data1 = (hydra_sine['l2_error_ls'],hydra_tanh['l2_error_ls'],don_sine['l2_error_ls'],don_tanh['l2_error_ls'])
data2 = (hydra_sine_reg['l2_error_ls'],hydra_tanh_reg['l2_error_ls'],don_sine_reg['l2_error_ls'],don_tanh_reg['l2_error_ls'])

positions = [0, 0.6, 1.2, 1.8]
i = 0
v1_knw_left, = ax.plot([-0.2+positions[i],0+positions[i]],[hydra_sine['knw_l2_error'],hydra_sine['knw_l2_error']],'r')
v1_knw_right, = ax.plot([0+positions[i],0.2+positions[i]],[hydra_sine_reg['knw_l2_error'],hydra_sine_reg['knw_l2_error']],'b')
i = 1
v2_knw_left, = ax.plot([-0.2+positions[i],0+positions[i]],[hydra_tanh['knw_l2_error'],hydra_tanh['knw_l2_error']],'r')
v2_knw_right, = ax.plot([0+positions[i],0.2+positions[i]],[hydra_tanh_reg['knw_l2_error'],hydra_tanh_reg['knw_l2_error']],'b')
i = 2
v3_knw_left, = ax.plot([-0.2+positions[i],0+positions[i]],[don_sine['knw_l2_error'],don_sine['knw_l2_error']],'r')
v3_knw_right, = ax.plot([0+positions[i],0.2+positions[i]],[don_sine_reg['knw_l2_error'],don_sine_reg['knw_l2_error']],'b')
i = 3
v4_knw_left, = ax.plot([-0.2+positions[i],0+positions[i]],[don_tanh['knw_l2_error'],don_tanh['knw_l2_error']],'r')
v4_knw_right, = ax.plot([0+positions[i],0.2+positions[i]],[don_tanh_reg['knw_l2_error'],don_tanh_reg['knw_l2_error']],'b')

v1 = ax.violinplot(data1, positions = positions, points=100,
               showmeans=False, showextrema=False, showmedians=False)

for b in v1['bodies']:
    # get the center
    m = np.mean(b.get_paths()[0].vertices[:, 0])
    # modify the paths to not go further right than the center
    b.get_paths()[0].vertices[:, 0] = np.clip(b.get_paths()[0].vertices[:, 0], -np.inf, m)
    b.set_color('r')
    
v2 = ax.violinplot(data2, positions = positions, points=100, 
               showmeans=False, showextrema=False, showmedians=False)

for b in v2['bodies']:
    # get the center
    m = np.mean(b.get_paths()[0].vertices[:, 0])
    # modify the paths to not go further left than the center
    b.get_paths()[0].vertices[:, 0] = np.clip(b.get_paths()[0].vertices[:, 0], m, np.inf)
    b.set_color('b')
    
ax.legend([v1['bodies'][0],v1_knw_left,v2['bodies'][0],v1_knw_right],['Vanilla Model',r'$\mathcal{K}$ Metric', r'Model + $\mathcal{K}$ Regularization',r'$\mathcal{K}$ Metric'], fontsize = 20, loc='upper left', bbox_to_anchor=(0.075, 1.15),ncol=4)
ax.set_yscale('log')
ax.set_ylabel(r'Relative $L^2$ Error', fontsize = 25)
ax.set_xticks(positions,['MH-PINN (sine)', 'MH-PINN (tanh)','PI-DON (sine)', 'PI-DON (tanh)'], fontsize = 20)
ax.tick_params(axis='y', which='major', length = 10, labelsize=15)
ax.yaxis.set_tick_params(labelsize=24)
ax.tick_params(axis='y', which='minor', length = 5)
plt.savefig(path + 'violinPlot_allencahn.png', dpi = 300, bbox_inches='tight')
```
